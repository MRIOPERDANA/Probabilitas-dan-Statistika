**1.41**

One way to compare the numbers of Facebook users for different regions of the world is to take into
account the populations of these regions. The market penetration for a product is the number of users
divided by the number of potential users, expressed as a percent. For Facebook, we use the population as
the number of potential users. Here are estimates of the populations in 2012 of the same geographic regions
that we studied in the previous exercise:

```python
import matplotlib.pyplot as plt

wilayah = ['Afrika', 'Timur Tengah', 'Asia', 'Amerika Utara', 'Karibia', 'Oseania/Australia', 'Amerika Tengah', 'Amerika Selatan', 'Eropa']
pengguna = [40, 20, 195, 173, 6, 14, 41, 113, 233]
populasi = [1026, 213, 3900, 347, 39, 36, 155, 402, 818]

penetrasi_pasar = [(pengguna[i] / populasi[i]) * 100 for i in range(len(wilayah))]

wilayah_sorted = [x for _, x in sorted(zip(penetrasi_pasar, wilayah), reverse=True)]
penetrasi_pasar_sorted = sorted(penetrasi_pasar, reverse=True)

plt.figure(figsize=(10, 6))
plt.bar(wilayah_sorted, penetrasi_pasar_sorted, color='blue')

plt.title('Penetrasi Pasar Facebook Menurut Wilayah (November 2012)')
plt.xlabel('Wilayah')
plt.ylabel('Penetrasi Pasar (%)')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

**B. pengecilan**

ada pengecilan ekstrem pada wilayah asia

```python
import matplotlib.pyplot as plt

wilayah = ['Afrika', 'Timur Tengah', 'Asia', 'Amerika Utara', 'Karibia', 'Oseania/Australia', 'Amerika Tengah', 'Amerika Selatan', 'Eropa']
pengguna = [40, 20, 195, 173, 6, 14, 41, 113, 233]
populasi = [1026, 213, 3900, 347, 39, 36, 155, 402, 818]


penetrasi_pasar = [(pengguna[i] / populasi[i]) * 100 for i in range(len(wilayah))]

plt.figure(figsize=(10, 6))
markerline, stemlines, baseline = plt.stem(wilayah, penetrasi_pasar, linefmt='grey', markerfmt='o', basefmt=' ')

plt.setp(markerline, color='blue', markersize=10)
plt.setp(stemlines, color='blue', linewidth=2)

plt.title('Penetrasi Pasar Facebook Menurut Wilayah (November 2012)')
plt.xlabel('Wilayah')
plt.ylabel('Penetrasi Pasar (%)')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
